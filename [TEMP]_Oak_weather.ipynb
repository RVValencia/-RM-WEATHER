```python
#Machine learning part
#WE try to predict tomorrow's temp_max using historical data
```

```python
#create a target
core_weather["target"] = core_weather.shift(-1)["temp_max"]
#shifts maximum temperature one row behind
```

```python
core_weather
```

```python
#get rid of nulls in last row
core_weather = core_weather.iloc[:-1,:].copy()
```

```python
core_weather
```

```python
#import
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
#create a list called predictors
predictors = ["precip", "temp_max", "temp_min"]
```

```python
#split into training set to test set
train = core_weather.loc[:"2020-12-31"]
```

```python
test = core_weather.loc["2021-01-01":]
```

```python
#fit method in regresison model (?)
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
#lets check how we did
from sklearn.metrics import mean_absolute_error
```

```python
#checks the accuracy

mean_absolute_error(test["target"], predictions)
```

```python
#On average, we are 0.07 off
```

```python
#what can we do to improve the accuracy (?)
```

```python
#combine our actual values and our prediction values
#we concatenate the actuals to the predictions
#the predictions are in numpy array, convert to panda series
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
```

```python
#(???) precipitation has negative impact on temp prediction, tempmax has the biggest impact, temp min has a smaller impact
```

```python
#let us define a function that does what we just did
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"], predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error, combined
```

```python
#rolling means

core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
```

```python
core_weather
```

```python
#NaNs in first rows because it takes 30 days before,
```

```python
#How different is the temperature that day from the monthly average
core_weather["month-day_max"] = core_weather["month_max"] / core_weather["temp_max"]
```

```python
#how different is the maximum temperature from the minimum temperature
core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"]
```

```python
#create a list called predictors adding the two above
predictors = ["precip", "temp_max", "temp_min" , "month_max", "month-day_max", "max_min"]
```

```python
#remove first the first 30 rows due to NaNs
core_weather = core_weather.iloc[30:,:].copy()
```

```python
#run function to get error and combined
error, combined = create_predictions(predictors, core_weather, reg)
```
